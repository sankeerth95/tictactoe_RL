```python
import numpy as np
import tictactoe_agent as ta
```

```python
def check_win(state, symbol='x'):
    win_states = [np.array([0, 1, 2]),\
                  np.array([3, 4, 5]),\
                  np.array([6, 7, 8]),\
                  np.array([0, 3, 6]),\
                  np.array([1, 4, 7]),\
                  np.array([2, 5, 8]),\
                  np.array([0, 4, 8]),\
                  np.array([2, 4, 6]),]

    ols = [ np.array_equal(state[y]==symbol, np.full(3, True)) for y in win_states ]
    return np.bitwise_or.reduce(ols)
```

```python
def check_draw(state):
    return np.array_equal(state=='.', np.full(9, False))
```

```python
qa1 = ta.QAgent('x')
qa2 = ta.QAgent('o')
```

```python
for i in range(10000):
    s1rew = 0.0
    s2rew = 0.0
    qa1.set_greed(0.1)
    qa2.set_greed(0.9)


    qa1.init_game()
    qa2.init_game()

    s1 = np.full(9, '.')
    while True:
        s2 = qa1.update_and_play(s1, 0)
        check_draw(s2)
        if check_win(s2, 'x'):
            s2rew = -1.0
            s1rew = 1.0
            break
        if check_draw(s2):
            s2rew = 0.1
            s1rew = 0.1
            break 
        s1 = qa2.update_and_play(s2, 0)
        if check_win(s1, 'o'):
            s2rew = 1.0
            s1rew = -1.0
            break

    qa1.update_final(s1rew)
    qa2.update_final(s2rew)
```

```python
init_state = np.array(['x', 'o', 'o', 'x', 'x', '.', 'o', '.', '.'])
qa1.get_best_afterstate(qa1.afterstates(init_state))
```
